```python
# python code to fetch dataset from mysql and save it into the dataframe df, use pandas and mysql.connector
import mysql.connector
import pandas as pd

try:
    # Establish a connection to the MySQL database
    cnx = mysql.connector.connect(
        host='127.0.0.1',
        port=3306,
        user='root',
        password='root',
        database='test_adda'
    )
    query = "SELECT * FROM user_events" 
    
    df = pd.read_sql(query, cnx) 
    cnx.close()
    print(df.head())
except mysql.connector.Error as err:
    if err.errno == 2003:
        print(f"Error: Cannot connect to MySQL server. Make sure it's running.")
    elif err.errno == 1045:
        print(f"Error: Invalid MySQL credentials.")
    else:
        print(f"MySQL Error: {err}")
except Exception as e:
    print(f"Unexpected error: {e}")
```
